**Initializing qubits with $\ket{00}$**<br>
**Adding Y gates changes the current state to $\ket{01}$ with a phase angle of $\pi$/2**<br>
Dirac Notation: (Y$\ket{0}$)$\ket{0}$  
**Adding H gate changes the current state to 1/$\sqrt{2}$($\ket{01}$ + $\ket{11}$) with phase angle $\pi$/2**<br>
Dirac Notation: $\ket{0}$(H$\ket{1}$)

**Initialize qubits with $\ket{01}$**<br>
**Adding CNOT gate has no change in the curcuit**<br>
**Apply X gate changes the current state to $\ket{11}$**<br>
Dirac notation: (X$\ket{0}$)$\ket{1}$<br>
**Applying another CNOT gate changes the current state to $\ket{01}$**<br>
Dirac notation: CX$\ket{11}$



```python
from qiskit import QuantumCircuit

circ = QuantumCircuit(2)
circ.y(0)
circ.i(1)
circ.barrier()
circ.i(0)
circ.h(1)
circ.measure_all()
circ.draw('mpl')
```

```python
circ1 = QuantumCircuit(2)
circ1.i(0)
circ1.x(1)
circ1.barrier()
circ1.cx(0,1)
circ1.barrier()
circ1.x(0)
circ1.i(1)
circ1.barrier()
circ1.cx(0,1)
circ1.measure_all()

circ1.draw('mpl')
```
