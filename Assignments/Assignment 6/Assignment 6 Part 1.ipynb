### Increment

Starting with $\ket{0}$ apply a X gate to increment the state to $\ket{1}$

```python
qc = QuantumCircuit(1) 
qc.reset(0)
qc.barrier()
qc.x(0)
qc.measure_all()
qc.draw('mpl')
```

Starting with $\ket{1}$ apply a X gate to increment the state to $\ket{0}$

```python
qc = QuantumCircuit(1) 
qc.reset(0)
qc.x(0)
qc.barrier()
qc.x(0)
qc.measure_all()
qc.draw('mpl')
```

### Decrement

Decrement in single qubit case is the same as increment. All we have to do is apply a X gate to the qubit state.

The following is the curcuit to increment any 2 qubit state

```python
qc = QuantumCircuit(2) 
qc.cnot(0,1)
qc.x(0)
qc.draw('mpl')
```

Increment $\ket{00}$ to $\ket{01}$

```python
qc = QuantumCircuit(2) 
qc.reset(0)
qc.reset(1)
qc.barrier()
qc.cnot(0,1)
qc.x(0)
qc.measure_all()
qc.draw('mpl')
```

Increment state $\ket{01}$ to $\ket{10}$

```python
qc = QuantumCircuit(2) 
qc.reset(0)
qc.reset(1)
qc.x(0)
qc.barrier()
qc.cnot(0,1)
qc.x(0)
qc.measure_all()
qc.draw('mpl')
```

Increment $\ket{10}$ to $\ket{11}$

```python
qc = QuantumCircuit(2) 
qc.reset(0)
qc.reset(1)
qc.x(1)
qc.barrier()
qc.cnot(0,1)
qc.x(0)
qc.measure_all()
qc.draw('mpl')
```

Increment $\ket{11}$ to $\ket{00}$

```python
qc = QuantumCircuit(2) 
qc.reset(0)
qc.reset(1)
qc.x(1)
qc.x(0)
qc.barrier()
qc.cnot(0,1)
qc.x(0)
qc.measure_all()
qc.draw('mpl')
```
